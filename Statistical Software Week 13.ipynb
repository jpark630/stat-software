```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#예제 1: y=x^2/10 그래프 그리기

x=np.arange(0,10)
y=(x**2)/10

plt.plot(x,y,color='blue')
plt.title("quadratic plot")
plt.xlabel("X")
plt.ylabel("Y")
```

```python
#예제2: sin(x) 함수 그래프 그리기

pi=3.14
x=np.arange(0,20)*pi/10
y=np.sin(x)

plt.plot(x,y,color='blue')
plt.title("sine plot")
plt.xlabel("x")
plt.ylabel("sin x")
```

```python
#그래프를 다른 디자인으로 그리기
sns.set()
pi=3.14
x=np.arange(0,20)*pi/10
y=np.sin(x)
plt.plot(x,y,color='blue')
plt.title("sine plot")
plt.xlabel("x"); plt.ylabel("sin x")
```

```python
#히스토그램 그리기

x=np.random.normal(size=100) #100개 난수
sns.histplot(x)
```

```python
plt.hist(x, bins=10, color='green') #y축: frequency
```

```python
plt.hist(x, density=True, color='pink') #y축: density
```

```python
x=np.random.normal(size=100)
plt.boxplot(x) #basic box plot
plt.show()
```

```python
#예제: A, B 두 그룹의 물고기 길이를 측정한 데이터

species=np.array(['A','A','A','A','A','B','B','B','B','B','B','B'])
length=np.array([2,3,4,5,6,5,6,7,7,8,8,9])
fish=pd.DataFrame({
    'species': species,
    'length': length
})
fish
```

```python
fish.groupby("species").describe()
```

```python
sns.boxplot(x='species',y='length',data=fish, color='green')
```
